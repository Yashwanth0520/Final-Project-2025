```python
import os as ana_ani_ops
import warnings as ana_ani_usrwrng
ana_ani_usrwrng.filterwarnings('ignore')

ana_animal_emopath = 'Analysing Spectrogram Parameters and Convolutional Neural Networks for Animal Emotion Recognition from Vocalizations/Code/'
ana_animal_emotion = ['Animal Anger', 'Animal Bur', 'Animal Disgust', 'Animal Fear']

# number of files in each folder
ana_animal_emocounts = {}

for ana_ani_emotion in ana_animal_emotion:
    emotion_folder = ana_ani_ops.path.join(ana_animal_emopath, ana_ani_emotion)
    if ana_ani_ops.path.exists(emotion_folder):
        ana_ani_numfiles = len([f for f in ana_ani_ops.listdir(emotion_folder) if f.endswith('.wav')])
        ana_animal_emocounts[ana_ani_emotion] = ana_ani_numfiles
    else:
        ana_animal_emocounts[ana_ani_emotion] = 0

# Display the number of audio files per emotion
for ana_ani_emotion, ecount in ana_animal_emocounts.items():
    print(f"{ana_ani_emotion}: {ecount} audio files")
```

```python
import librosa as ana_animal_Elbr
import librosa.display as ana_ani_lbrdisp
import matplotlib.pyplot as ana_animalplt
import IPython.display as ana_animalipd
import random as ana_animalRand

# Select a random audio file from one of the emotion folders
ana_animal_exemotion = ana_animalRand.choice(ana_animal_emotion)
ana_animal_exfolder = ana_ani_ops.path.join(ana_animal_emopath, ana_animal_exemotion)

ana_animal_exaudio_fil = ana_animalRand.choice([f for f in ana_ani_ops.listdir(ana_animal_exfolder) if f.endswith('.wav')])
ana_animal_exaudio_pat = ana_ani_ops.path.join(ana_animal_exfolder, ana_animal_exaudio_fil)

# Load the emotion audio
ana_ani_y, ana_ani_sampr = ana_animal_Elbr.load(ana_animal_exaudio_pat, sr=None)

print(f"Playing Animal Emotions Audio from {ana_animal_exemotion}: {ana_animal_exaudio_fil}")
ana_animalipd.Audio(ana_animal_exaudio_pat)
```

```python
# Distribution of audio files per emotion
ana_animalplt.figure(figsize=(7, 4))
ana_animalplt.bar(ana_animal_emocounts.keys(), ana_animal_emocounts.values(), color=['lightgreen', 'pink', 'cyan', 'lavender'], edgecolor='black')
ana_animalplt.xlabel("Animal Emotions", fontsize=12)
ana_animalplt.ylabel("Number of Emotion Audio Files", fontsize=12)
ana_animalplt.title("Distribution of Audio Files in Animal Emotions\n", fontsize=14, fontweight='bold')
ana_animalplt.xticks(rotation=20)
ana_animalplt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the count on top of each bar
for ei, ecount in enumerate(ana_animal_emocounts.values()):
    ana_animalplt.text(ei, ecount + 1, str(ecount), ha='center', fontsize=12, fontweight='bold')

ana_animalplt.show()
```
